```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
df0 = pd.read_excel("Dry_Bean_Dataset.xlsx")
```

```python
beans = set(df0["Class"])
print(beans)
one_hot = pd.get_dummies(df0['Class'], drop_first=False)
df1 = pd.concat([df0, one_hot], axis=1)
df1.head()
```

```python
df2train, df2test = train_test_split(df1, test_size = .25, random_state = 12345)
fill, df2test_results = train_test_split(df1, test_size = .25, random_state = 12345)
```

```python
df2test_results
```

```python
mlp = MLPClassifier(solver='adam', hidden_layer_sizes=50,max_iter=100000, shuffle=False, random_state=12345)
for x in beans:
    mlp.fit(df2train.drop(["Class", 'SEKER', 'HOROZ', 'CALI', 'DERMASON', 'BARBUNYA', 'SIRA', 'BOMBAY'],
                      axis = 1), df2train[x])
    df2test_results[x] = mlp.predict(df2test.drop(["Class", 'SEKER', 'HOROZ', 'CALI', 'DERMASON', 'BARBUNYA', 'SIRA', 'BOMBAY'],
                                     axis = 1))
```

```python
df3 = df2test_results
    
```

```python
data = np.empty((8,7))
y = 0
for x in beans:
    #print()
    #print(x)
    data[0,y] = df3[df3["Class"] == x].shape[0]
    #df3.columns
    z = 0
    for a in beans:
        z += 1
        #print(a)
        data[z,y] = df3[df3["Class"] == x].sum(axis = 0)[a]
    y += 1
        
print(pd.DataFrame(data, columns = ['DERMASON', 'SEKER', 'SIRA', 'HOROZ', 'BARBUNYA', 'BOMBAY', 'CALI'],
                   index = ["size",'DERMASON', 'SEKER', 'SIRA', 'HOROZ', 'BARBUNYA', 'BOMBAY', 'CALI']))
        
        
```
